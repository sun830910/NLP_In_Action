# 4种聊天机器人的实现方法
## 模式匹配方法
优点：  
1. 易于上手
2. 训练易于复用
3. 模块化
4. 易于控制/约束
  
缺点：
1. 有限的领域
2. 能力受到人力投入的限制
3. 调试困难
4. 硬性脆弱规则

## 基础方法
优点：  
1. 擅长回答逻辑问题
2. 易于控制/约束  
  
缺点：
1. 听起来像是人工的、机械的
2. 难以处理含糊不清的情况
3. 难以处理尝试
4. 受结构化数据的限制
5. 需要大规模的信息抽取
6. 需要人工管理

## 搜索方法
优点：  
1. 简单
2. 容易训练
3. 可以模仿人的对话
  
缺点：  
1. 难以扩展
2. 不连贯的人格  
3. 无法感知上下文  
4. 无法回答事实性问题

## 生成方法
优点：  
1. 新的、富有创造力的对话方式
2. 更少的人力投入
3. 领域仅受数据限制
4. 上下文感知
  
缺点：
1. 难以引导
2. 难以训练
3. 需要跟多的对话数据
4. 需要更多的处理才能训练
