# 信息提取
将非结构化文本转为存储在知识库或知识图谱中的结构化信息。  
# 正则模式

```python
# 硬编码在Python代码中的模式
def find_greeting(s):
    if s[0] == 'H':
        if s[:3] in ['Hi', 'Hi ', 'Hi,', 'Hi!']:
            return s[:2]
        elif s[:6] in ['Hello', 'Hello ', 'Hello,', 'Hello!']:
            return s[:5]
    elif s[0] == 'Y':
        if s[1] == 'o' and s[:3] in ['Yo', 'Yo ', 'Yo,', 'Yo!']:
            return s[:2]
    return None
```

```python
find_greeting("Hi Mr.Turing!")
```

```python
find_greeting("Hello, Rosa.")
```
