# 正则匹配

```python
import re
```

```python
patterns = "(hi|hello|hey)[ ]*([a-z]*)"
re.match(patterns, "Hello Rosa", flags=re.IGNORECASE)  # 为使正则表达式更简单，通常忽略文本字符的大小写
```

```python
re.match(patterns, "hi ho, hi ho, it's off to work...", flags=re.IGNORECASE)
```

```python
re.match(patterns, "hey, what's up", flags=re.IGNORECASE)
```

在正则表达式中，可以使用方括号指定某个字符类，还可以用短横线表示范围而不需要逐个输入，星号则表示为匹配任意数量的属于该字符类的连续字符。  
下面将正则写的更细致以匹配更多的问候语

```python
patt = r"[^a-z]*([y]o|[h']?ello|ok|hey|(good[ ])?(morn[gin']{0,3}|afternoon|even[gin']{0,3}))[\s,;:]{1,3}([a-z]{1,20})"
```

引号前的"r"指定的是一个原始字符串，而不是正则表达式，使用Python原始字符串可以将反斜杠直接传递给正则表达式编译器，无需在所有特殊的正则表达式字符前面加双反斜杠，这些特殊字符包括空格和{}。

```python
re_gretting = re.compile(patt, flags=re.IGNORECASE)  # 编译正则表达式，这样就不用在每次使用他们时指定选项
```

```python
re_gretting.match("Hello Rosa")
```

```python
re_gretting.match("Hello Rosa").groups()
```

```python
re_gretting.match("Good morning Rosa")
```

```python
re_gretting.match("Good morning Rosa").groups()
```

```python
re_gretting.match("Good evening Rosa Parks").groups()
```

```python
re_gretting.match("Good Morn'n Rosa")
```

```python
re_gretting.match("yo Rosa")
```

```python
re_gretting.match("Good Manning Rosa")
```

## 通过添加一个输出生成器最终得到一个只用一种技巧的聊天机器人。

```python
my_names = set(['rosa', 'rose', 'chatty', 'chatbot', 'bot', 'chatterbot'])
curt_names = set(['hal', 'you', 'u'])
greeter_name = ''

match = re_gretting.match(input())
if match:
    at_name = match.groups()[-1]
    if at_name in curt_names:
        print("Good one.")
    elif at_name.lower() in my_names:
        print("Hi {}, How are you?".format(greeter_name))
```
